## ques4

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

r = Rectangle(10, 5)

print("Area of Rectangle:", r.area())
```

## ques5

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

c = Circle(7)

print("Area of Circle:", c.area())
```

## ques6

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Balance after deposit:", self.balance)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Balance after withdrawal:", self.balance)
        else:
            print("Insufficient Balance")

acc = BankAccount("Nikhil", 10000)

acc.deposit(5000)
acc.withdraw(3000)
```

## ques8

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, roll_number):
        super().__init__(name, age)
        self.roll_number = roll_number

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll Number:", self.roll_number)

s = Student("Nikhil", 21, 42)

s.display()
```

## ques9

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b != 0:
            return a / b
        return "Cannot divide by zero"

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

c = Calculator()

print("Addition:", c.add(num1, num2))
print("Subtraction:", c.subtract(num1, num2))
print("Multiplication:", c.multiply(num1, num2))
print("Division:", c.divide(num1, num2))
```

## ques10

```python
class LibraryBook:
    def __init__(self, book_name, author, price):
        self.book_name = book_name
        self.author = author
        self.price = price

    def display_book_info(self):
        print("Book Name:", self.book_name)
        print("Author:", self.author)
        print("Price:", self.price)
        print("-" * 30)

book1 = LibraryBook("Python Programming", "Guido van Rossum", 500)
book2 = LibraryBook("Data Science", "Joel Grus", 650)
book3 = LibraryBook("Machine Learning", "Tom Mitchell", 700)

book1.display_book_info()
book2.display_book_info()
book3.display_book_info()
```
