## ques6

```python
arr = [10, 20, 30, 40, 50, 60]

def reverse_list(lst):
    rev = []
    for i in range(len(lst)-1, -1, -1):
        rev.append(lst[i])
    return rev

print("Original List:", arr)
print("Reversed List:", reverse_list(arr))
```

## ques7

```python
data = (5, 10, 15, 20, 25, 30, 35)

count = 0
total = 0

for num in data:
    if num % 5 == 0:
        count += 1
    total += num

average = total / len(data)

print("Elements divisible by 5:", count)
print("Sum of all elements:", total)
print("Average:", average)
```

## ques8

```python
students = {
    "Aman": 78,
    "Riya": 92,
    "Kriti": 88,
    "Rahul": 95
}

highest = max(students, key=students.get)
lowest = min(students, key=students.get)

print("Highest Marks:")
print(highest, "->", students[highest])

print("\nLowest Marks:")
print(lowest, "->", students[lowest])

print("\nStudents scoring more than 85 marks:")
for name, marks in students.items():
    if marks > 85:
        print(name, "->", marks)
```

## ques9

```python
def count_frequency(arr):
    freq = {}

    for item in arr:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    for key, value in freq.items():
        print(key, "->", value, "times")

arr = [1, 2, 2, 3, 1, 4, 2]
count_frequency(arr)
```

## ques10

```python
def find_duplicates(arr):
    duplicates = []

    for item in arr:
        if arr.count(item) > 1 and item not in duplicates:
            duplicates.append(item)

    print("Duplicate elements are:")
    for item in duplicates:
        print(item)

arr = [10, 20, 30, 20, 40, 10, 50, 30]
find_duplicates(arr)
```
